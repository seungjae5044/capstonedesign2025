```python
import os
import soundfile as sf  # 음성 파일을 읽기 위한 라이브러리
from vosk import Model, KaldiRecognizer
from pydub import AudioSegment  # 오디오 변환 라이브러리

# VOSK 모델 경로 설정 (다운로드한 모델 폴더 경로)
MODEL_PATH = "vosk-model-small-ko/vosk-model-small-ko-0.22"
if not os.path.exists(MODEL_PATH):
    print("모델 경로가 잘못되었습니다. 한국어 모델을 다운로드하세요.")
    exit(1)

# VOSK 모델 로드
model = Model(MODEL_PATH)
recognizer = KaldiRecognizer(model, 16000)  # 샘플링 레이트는 16kHz로 설정

def convert_to_16k(file_path):
    """샘플링 레이트를 16kHz로 변환"""
    print(f"샘플링 레이트를 16kHz로 변환 중: {file_path}")
    audio = AudioSegment.from_file(file_path)
    audio = audio.set_frame_rate(16000)  # 샘플링 레이트를 16kHz로 설정
    audio = audio.set_channels(1)  # 모노 채널로 변환
    converted_path = "converted_16k.wav"
    audio.export(converted_path, format="wav")
    print(f"변환 완료: {converted_path}")
    return converted_path

def process_audio_file(file_path):
    """음성 파일을 실시간 처리"""
    if not os.path.exists(file_path):
        print("음성 파일 경로가 잘못되었습니다.")
        return

    # 음성 파일 읽기
    audio_data, samplerate = sf.read(file_path, dtype="int16")
    if samplerate != 16000:
        print("샘플링 레이트가 16kHz가 아닙니다. 변환을 진행합니다...")
        file_path = convert_to_16k(file_path)  # 파일 변환
        audio_data, samplerate = sf.read(file_path, dtype="int16")

    print("음성 파일을 처리합니다...")
    for i in range(0, len(audio_data), 16000):  # 데이터를 작은 블록으로 나누어 처리
        block = audio_data[i:i + 16000].tobytes()
        if recognizer.AcceptWaveform(block):
            result = recognizer.Result()
            print("인식 결과:", result)
        else:
            partial_result = recognizer.PartialResult()
            print("부분 결과:", partial_result)

if __name__ == "__main__":
    # 여기에 음성 파일 경로를 입력하세요
    AUDIO_FILE_PATH = "data1.wav"  # 처리할 음성 파일 경로
    process_audio_file(AUDIO_FILE_PATH)
```
